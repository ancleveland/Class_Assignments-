```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np

# Create the X and Y axis lists
time = np.arange(0,130,10)
dd = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rg = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]

# Plot the charts and apply some styling

danger_drop, = plt.plot(time, dd, color="blue", linewidth=1, label="Danger Drop")
rail_gun, = plt.plot(time,rg, color="Red", linewidth=1, label="Rail Gun")

# Add labels to X and Y axes :: Add title

plt.title("Coaster Speed")
plt.xlabel("Coaster Run Time")
plt.ylabel("Coaster Speed (MPH)")

# Set the limits for the X and Y axes
plt.xlim(0,120)
plt.ylim(5,95)

# Create a legend for the chart
plt.legend(handles=[danger_drop, rail_gun], loc="best")

# Add in a grid for the chart
plt.grid()

plt.show()
```
